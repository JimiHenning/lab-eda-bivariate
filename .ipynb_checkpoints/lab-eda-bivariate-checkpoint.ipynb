```python
df = pd.read_csv("amz_uk_price_prediction_dataset.csv")
df
```

```python
crosstab = pd.crosstab(df['category'], df['isBestSeller'])

crosstab.sort_values(by=True, ascending = False)
```

The most absolute bestsellers are found in the Grocery, Care, Baby, Sports and Tools categories. Now to calculate the proportion:

```python

crosstab_proportion = pd.crosstab(df['category'], df['isBestSeller'], normalize='index')


crosstab_proportion.sort_values(by=True, ascending=False)
```

Here it is evident that the most bestseller compared to the number of products are found in the Grocery, Lighting, Care, Phone Accessories and Tools categories, with Baby and Sports not being in the top spots anymore due to having less bestsellers relatively to the total products offered.

```python
crosstab
```

```python
from scipy.stats import chi2_contingency

# Chi-square test for 'MSZoning' and 'SaleCondition'
chi2_statistic, chi2_p_value, _, _ = chi2_contingency(crosstab)

chi2_statistic, chi2_p_value
```

A larger chi2 statistic value indicates a larger difference between what was observed and what would be expected if the variables were independent, here it is very high indicating a strong association. THe p-value here is displayed as 0, meaning it is probably so small that it get's rounded to 0, also indicating a statistically significant relationship between product type and bestseller status.

```python
#now to check for strength of association we do the cramer's V test
from scipy.stats.contingency import association

# Computing the association between variables in 'crosstab_result' using the "cramer" method
association(crosstab, method="cramer")
```

Though we had such clear chi2 results, the Cramér's V value of 0.1222829439760564 suggests a weak association. Though statistically significant, the strength of this relationship is not very strong in practical terms

```python
#Removing outliers from the product prices:
def turkeys_test_outliers(my_data, column_name, method="show"):
    # Ensure that the input is a DataFrame and access specific column
    data = my_data[column_name].copy()
    
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    
    # Define bounds for the outliers
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Identify the outliers
    outliers = (data < lower_bound) | (data > upper_bound)

    if method == "show":
        return data[outliers]  # This shows the outliers
    elif method == "replace":
        median = data.median()
        data[outliers] = median  # Replace outliers with median
        my_data[column_name] = data  # Update the DataFrame
        return my_data  # Return modified DataFrame
    elif method == "delete":
        # Filter the DataFrame to keep non-outlier rows
        data_no_outliers = my_data[~outliers]  # Keep rows that are not outliers
        return data_no_outliers  # Return DataFrame without outliers
```
